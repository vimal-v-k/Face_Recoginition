```python
import os
import cv2
import numpy as np
import face_recognition as fr
```

```python
path = 'images'
```

```python
mylist = os.listdir(path)
mylist
```

```python
imgs = []
classnames = []
```

```python
for img in mylist:
    imgpath = os.path.join(path,img)
    # print(imgpath)
    current_img = cv2.imread(imgpath)
    # print(current_img)
    imgs.append(current_img)
    classnames.append(img.split('.')[0])
print(classnames)
```

```python
img = imgs[0]

# if img is None:
#     raise ValueError("Image not loaded properly! Check the file path or source.")

img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# img = img.copy()
# img = np.array(img, dtype=np.uint8)
print("Image dtype:", img.dtype)  # Should be 'uint8'
print("Image shape:", img.shape) 
# # Display the image
# plt.imshow(img)
# plt.axis("off")
# plt.show()

face_in_frame = fr.face_locations(img)
# print(face_in_frame)
# img = np.array(img, dtype=np.uint8)
print(face_in_frame)
encode = fr.face_encodings(img,face_in_frame)[0]
print(encode)
```

```python
def faceencodings(images):
    encoding_list = []
    for i in imgs:
        img = cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        if face_in_frame:
            encode = fr.face_encodings(img, face_in_frame)[0]  # Fix: Pass `img` as the first argument
            encoding_list.append(encode)
        else:
            print("⚠️ No face detected in an image, skipping encoding.")

    return encoding_list
```

```python
encodelist_knownfaces=faceencodings(imgs)
encodelist_knownfaces
```

```python
video = cv2.VideoCapture(0)
while True:
    syc,frame = video.read()
    frame1 = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face_in_frame = fr.face_locations(frame1)
    # print(face_in_frame)
    encoded_face = fr.face_encodings(frame1,face_in_frame)
    # print(encoded_face)
    for enc_face,faceloc in zip(encoded_face,face_in_frame):
        matches = fr.compare_faces(encodelist_knownfaces,enc_face)
        print(matches)

        face_dis = fr.face_distance(encodelist_knownfaces,enc_face)
        print(face_dis)

        match_indx = np.argmin(face_dis)
        if matches[match_indx]:
            name = classnames[match_indx]
            # print(match_indx)
        else:
            name = "Unknown" 
        
        y1,x2,y2,x1 = faceloc
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),2)
        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,0),2)

    cv2.imshow("face",frame)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
